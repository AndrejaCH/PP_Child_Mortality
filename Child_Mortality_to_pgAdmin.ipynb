Import Dependencies

```python
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

read in datasets

```python
table01 = pd.read_csv("resources/AcuteLowerRespiratoryInfections.csv")
table02 = pd.read_csv("resources/BirthAphyxiaAndBirthTrauma.csv")
table03 = pd.read_csv("resources/CongenitalAnomalies.csv")
table04 = pd.read_csv("resources/DiarrhoealDiseases.csv")
table05 = pd.read_csv("resources/HIV-AIDS.csv")
table06 = pd.read_csv("resources/Injuries.csv")
table07 = pd.read_csv("resources/Malaria.csv")
table08 = pd.read_csv("resources/Measles.csv")
table09 = pd.read_csv("resources/MeningitisAndEncephalitis.csv")
table10 = pd.read_csv("resources/OtherCommunicablePerinatalandNutritionalConditions.csv")
table11 = pd.read_csv("resources/OtherNoncommunicableDiseases.csv")
table12 = pd.read_csv("resources/Prematurity.csv")
table13 = pd.read_csv("resources/SepsisAndOtherInfections.csv")
```

```python
table01.head(5)
```

```python
merged01 = pd.merge(table01, table02, how="left", on=["Year", "Age", "Country"])
merged02 = pd.merge(table03, table04, how="left", on=["Year", "Age", "Country"])
merged03 = pd.merge(table05, table06, how="left", on=["Year", "Age", "Country"])
merged04 = pd.merge(table07, table08, how="left", on=["Year", "Age", "Country"])
merged05 = pd.merge(table09, table10, how="left", on=["Year", "Age", "Country"])
merged06 = pd.merge(table11, table12, how="left", on=["Year", "Age", "Country"])
```

```python
merged07 = pd.merge(merged01, merged02, how="left", on=["Year", "Age", "Country"])
merged08 = pd.merge(merged03, merged04, how="left", on=["Year", "Age", "Country"])
merged09 = pd.merge(merged05, merged06, how="left", on=["Year", "Age", "Country"])
```

```python
merged10 = pd.merge(merged07, merged08, how="left", on=["Year", "Age", "Country"])
merged11 = pd.merge(merged09, table13, how="left", on=["Year", "Age", "Country"])
```

```python
child_mortality_df = pd.merge(merged10, merged11, how="left", on=["Year", "Age", "Country"])
```
